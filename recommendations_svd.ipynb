```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
ru_ids = pd.read_csv('data/all_ru_ids_games_20min_mean+try2.csv')
ru_ids
```

```python
ss_scaler = StandardScaler()
ru_ids.iloc[:,:-1] = ss_scaler.fit_transform(ru_ids.iloc[:,:-1])
ru_ids
```

```python
A = ru_ids / np.sqrt(ru_ids.shape[0] - 1)
U, S, V = np.linalg.svd(A)

#Function to calculate the cosine similarity (sorting by most similar and returning the top N)
def top_cosine_similarity(data, app_id, top_n=10):
    index = app_id 
    app_row = data[index, :]
    magnitude = np.sqrt(np.einsum('ij, ij -> i', data, data))
    similarity = np.dot(app_row, data.T) / (magnitude[index] * magnitude)
    sort_indexes = np.argsort(-similarity)
    return sort_indexes[:top_n]
```

```python
k = 50
app_id =631 # (вставляешь номер столбца с игрой)
top_n = 10
sliced = V.T[:, :k] # representative data
indexes = top_cosine_similarity(sliced, app_id, top_n)
indexes #выдает индексы столбцов с игрой
```
